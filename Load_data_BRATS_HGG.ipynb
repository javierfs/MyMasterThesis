```python
from PIL import Image
import scipy
from scipy import misc
import numpy as np
import os 
import random
import shutil
import os.path
import nibabel
from sklearn.preprocessing import MinMaxScaler
#Only considering HGG due to its larger size of tumour
path_data = '/Users/Sanchezj/Brats17TrainingData/HGG'
outpath_data = '/Users/Sanchezj/preprocess_Data/'
files = os.listdir(path_data)
nfolders = len(files)
#num_to_select = len(files) #Number is 210
#random.seed(9001)
#list_of_random_files = random.sample(files, num_to_select)
#print('This is number of files: ' + str(num_to_select ))

        #if thing in some_list: some_list.remove(thing)
#Brats17_2013_2_1_flair.nii.gz
#total_i = 800
#i=0
nfiles = 0
dct_img_slcs = {}
nfiles_seg = 0;
nfiles_flair = 0;

for name_folder in files: 
    if name_folder.startswith("Brats17"):
        #nfiles+=1
        path_folder = path_data+'/'+name_folder+'/'
        MRI_types_scan = os.listdir(path_folder)
        
        for file in MRI_types_scan:
            slices_list = []
            if  file.startswith("Brats17") and  file.endswith("seg.nii.gz"):
                #print(file)
                nfiles_seg = nfiles_seg + 1
                path2finalfile_seg = path_folder+file
                img_seg = nibabel.load(path2finalfile_seg)
                data = img_seg.get_data()
                
                for i in range(data.shape[2]): #is it 155?
                    slice_i = data[:, :, i]
                    if np.std(slice_i) > 0: #remove those slices with 0 variance
                        slices_list.append(i)
                        
                        slice_normalized = (slice_i - np.mean(slice_i)).astype(np.float32)/(np.std(slice_i)).astype(np.float32)
                        slice_clipped = np.clip(slice_normalized, -5, 5)
                        scaler = MinMaxScaler()
                        slice_final_normalization = scaler.fit_transform(slice_clipped)
                        save_filename = file.replace('.nii.gz', '')
                        scipy.misc.imsave(outpath_data+str(i)+save_filename+ '_' +'.jpg', slice_final_normalization)
                        
                #store filenames pair-segm with the slices with non-zero variance
                #print(file)
                img_name = path2finalfile_seg.replace('seg.nii.gz','')
                
                full_pair_name = img_name+'flair.nii.gz'
                #print(slices_list)
                nfiles_flair = nfiles_flair + 1
                img_flair = nibabel.load(full_pair_name)
                data_flair = img_flair.get_data()
                for k in slices_list: 
                    slice_k = data_flair[:, :, k]
                    slice_normalized_k = (slice_k - np.mean(slice_k)).astype(np.float32)/(np.std(slice_k)).astype(np.float32)
                    slice_clipped_k = np.clip(slice_normalized_k, -5, 5)
                    scaler = MinMaxScaler()
                    slice_final_normalization_k = scaler.fit_transform(slice_clipped_k)
                    #full_pair_name = full_pair_name.replace('.nii.gz', '')
                    
                    save_filename = file.replace('seg.nii.gz', 'flair')
                    scipy.misc.imsave(outpath_data+str(k)+save_filename+ '_' +'.jpg', slice_final_normalization_k)
                
                
                    
                    
print('#img flair: ' + str(nfiles_flair))               
print('#img seg: ' + str(nfiles_seg))
    

   
    
    
    
```
